```python
import pandas as pd

data["Timestamp"] = pd.to_datetime(
    data["Date"] + " " + data["Time"],
    format="%m/%d/%Y %H:%M:%S",
    errors="coerce"  
)

data = data.dropna(subset=["Timestamp"])
data["Timestamp"] = data["Timestamp"].astype("int64") // 10**9

final_data = data.drop(["Date", "Time"], axis=1)

final_data.head()
```
