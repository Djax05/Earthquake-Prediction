```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
data = pd.read_csv('../data/database.csv')
```

```python
data.info()
```

```python
data = data[["Date", "Time", "Latitude","Longitude", "Type", "Depth", "Magnitude"]]
```

```python
data.head()
```

```python
timestamp = []

def combine_to_datetime(data: pd.DataFrame):

    for i, j in zip(data["Date"], data["Time"]):
        if j == i:
            timestamp.append(i)
            continue
        dateandtime = f"{i}  {j}"
        try:
            datetime = pd.to_datetime(dateandtime, format="mixed")
        except Exception as e:
            print(dateandtime, "->", e)
        timestamp.append(datetime)
    return timestamp



combine_to_datetime(data)
data["Timestamp"] = timestamp
```
